```python
# Importing Dependencies
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_Super_Bowl_starting_quarterbacks"
```

```python
table = pd.read_html(url)
table
```

```python
sb_df = table[1]
sb_df.head(5)
```

```python
# Renaming 'Team' Columns
sb_df = sb_df.rename(columns={"Team":"Winning Team", "Team.1":"Losing Team"})
sb_df.head(5)
```

```python
sb_df.to_csv("Resources/sb_df.csv", index=False)
```

```python
# Reading back in the cleaned csv
file = "Resources/sb_df_cleaned.csv"

sb_df_cleaned = pd.read_csv(file)  
sb_df_cleaned.head(5)
```

```python
sb_df_cleaned = sb_df_cleaned.rename(columns={"Winning QB":"Quarterback", "Winning Team":"Team"})
sb_df_cleaned.head(5)
```

```python
# Fixing 'Season' Column
sb_df_cleaned["Season"] = sb_df_cleaned["Season"].str.replace("1966\u2009AFL/NFL","1966")
sb_df_cleaned["Season"] = sb_df_cleaned["Season"].str.replace("1967\u2009AFL/NFL","1967")
sb_df_cleaned["Season"] = sb_df_cleaned["Season"].str.replace("1968\u2009AFL/NFL","1968")
sb_df_cleaned["Season"] = sb_df_cleaned["Season"].str.replace("1969\u2009AFL/NFL","1969")
sb_df_cleaned.head(5)
```

```python
# Fixing 'Winning QB' Column
sb_df_cleaned["Quarterback"] = sb_df_cleaned["Quarterback"].str.replace("*","")
sb_df_cleaned["Quarterback"] = sb_df_cleaned["Quarterback"].str.replace("‡","")
sb_df_cleaned["Quarterback"] = sb_df_cleaned["Quarterback"].str.replace("MVP","")
sb_df_cleaned.head(5)
```

```python
# Fixing 'Winning Team' & 'Losing Team' Columns
sb_df_cleaned["Team"] = [x[:-1] for x in sb_df_cleaned["Team"]]
sb_df_cleaned.head(5)
```

```python
# Dropping last row as it looks to be a repeat header row
sb_df_cleaned = sb_df_cleaned.drop([108])
sb_df_cleaned = sb_df_cleaned.sort_values(["Season"])
sb_df_cleaned = sb_df_cleaned.set_index(["Season"])
sb_df_cleaned
```

```python
# Reading third table from website - refers to Superbowl starts by QB and win percentages
qb_sb_starts_df = table[2]
qb_sb_starts_df.head(5)
```

```python
# Dropping Superbowl and Reference columns
qb_sb_starts_df = qb_sb_starts_df.drop(columns=["Super Bowl(s)", "Ref"])
qb_sb_starts_df = qb_sb_starts_df.drop([25])
qb_sb_starts_df.head(5)
```

```python
# Saving to CSV to make some quick formatting edits
qb_sb_starts_df.to_csv("Resources/qb_sb_starts_df.csv", index=False)
```
